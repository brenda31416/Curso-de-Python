```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt


def funcion_der(cadena):
    x = sp.symbols('x')
    fig, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True)
    x_valores = np.linspace(-10, 10, 100)
    expr = sp.sympify(cadena)
    f = sp.lambdify(x, expr, 'numpy')
    ax1.set_title("Funcion")
    ax1.plot(x_valores, f(x_valores))
    deriv_expr = sp.diff(expr, x)
    deriv_f = sp.lambdify(x, deriv_expr, 'numpy')
    ax2.set_title("Derivada")
    ax2.plot(x_valores, deriv_f(x_valores))
    plt.show()

funcion_der("x**2")
```
