Consideraremos el problema de clasificar todas las ternas pitag√≥ricas (x,y,z) donde (x<=1000,y<=1000,z<=1000)

```python

def es_potencia(n,pot=2):
    return int(pow(n,1/pot)) ** pot == n

ternas_pit=[]
for i in range(1,101):
    x=i**2
    for j in range(i,1001):
        y=j**2
        if es_potencia(x+y):
            z = int(sqrt(x+y))
            ternas_pit.append((i,j,z))
ternas_pit
```
